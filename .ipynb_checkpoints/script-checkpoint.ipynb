```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import OrdinalEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
from sklearn.metrics import roc_curve
```

```python
df = pd.read_csv('train.csv')
```

```python
print(df.head())
#y_train = df['Survived']
X_train = df.drop(['PassengerId', 'Name', 'Ticket'],axis=1)
```

```python
print(X_train.info())
```

```python
print(X_train['Embarked'].value_counts())
```

```python
print(X_train.describe())
```

```python
X_train.hist(bins=30, figsize=(20,15))
plt.show()
```

```python
sex_col = X_train[['Sex']]
ordinal_encoder = OrdinalEncoder()
sex_encoded = ordinal_encoder.fit_transform(sex_col)
X_train['sex_encoded'] = sex_encoded
X_train.drop(['Sex', 'Cabin'], axis=1, inplace=True) #Cabin may be useful
```

```python
median = X_train['Age'].median()
X_train['Age'].fillna(median,inplace=True)
```

```python
X_train.dropna(inplace=True)
```

```python
X_train.info()
```

```python
oh_encoder = OneHotEncoder(sparse=False)
ohe_columns = oh_encoder.fit_transform(X_train[['Embarked']])
```

```python
df_embarked = pd.DataFrame(ohe_columns, columns=['C', 'Q', 'S'])
```

```python
X_train_n = np.c_[X_train, df_embarked]
```

```python
X_train_n = pd.DataFrame(X_train_n, columns=list(X_train.columns) + list(df_embarked.columns), index=X_train.index)
```

```python
X_train_n.drop(['Embarked'],axis=1,inplace=True)
```

```python
X_train_n = X_train_n.apply(pd.to_numeric)
print(X_train_n)
```

```python
corr_mx = np.array(X_train_n.corr())
```

```python
print(np.array(corr_mx))
```

```python
np.fill_diagonal(corr_mx, 0)
```

```python
plt.matshow(corr_mx, cmap=plt.cm.gray)
plt.show()
```

```python
X_train_n.info()
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_n.astype(np.float64))
```

```python
X_train_scaled
```

```python
knn_clf = KNeighborsClassifier()
```

```python
y_train = X_train_n['Survived']
```

```python
knn_clf.fit(X_train_scaled, y_train)
```

```python
a = cross_val_score(knn_clf, X_train_scaled, y_train, cv=3, scoring='accuracy')
```

```python
print(a)
```

```python
y_train_pred = cross_val_predict(knn_clf, X_train_scaled, y_train, cv=3)
```

```python
y_scores = cross_val_predict(knn_clf, X_train_scaled, y_train, cv=3, method="predict_proba")
```

```python
confusion_matrix(y_train, y_train_pred)
```

```python
precision_score(y_train, y_train_pred)
```

```python
recall_score(y_train, y_train_pred)
```

```python
f1_score(y_train, y_train_pred)
```

```python
from sklearn.metrics import precision_recall_curve
y_scores = y_scores[:, 1]
precisions, recalls, thresholds = precision_recall_curve(y_train, y_scores)
```

```python
fpr, tpr, thresholds = roc_curve(y_train, y_scores)
```
